```python
from math import comb

def compare_sum_comb_and_pow(n, m):
    sum_comb = sum(comb(n, k) for k in range(1, m+1))
    pow_nm = n ** m
    return sum_comb, pow_nm, sum_comb / pow_nm if pow_nm else float('inf')

n, m = 20, 10
sum_comb, pow_nm, ratio = compare_sum_comb_and_pow(n, m)
print(f"Сумма биномиальных: {sum_comb}, n^m: {pow_nm}, отношение: {ratio}")
```
